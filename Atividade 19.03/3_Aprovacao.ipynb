```python
nota = int(input("Insira o valor da sua nota -->"))
frequency = int(input("Insira a porcentagem de aulas presenciadas -->"))
```

```python
if nota >= 7:
    if frequency >= 75:
        print("Você foi *APROVADO*")
    else:
        print("REPROVADO por falta")
else:
    print("REPROVADO por baixa nota")
```
