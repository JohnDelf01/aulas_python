```python
num = int(input("Digite um número inteiro-->"))
```

```python
if num > 0:
    if num != 1:
        if num % 2 == 0:
            print("Número Múltiplo de 2")
        if num % 3 == 0:
            print("Número Múltiplo de 3")
    else:
        print("O número '1' não possui múltiplos")
elif num == 0:
    print("Número '0' não possui múltiplos")
else:
    print("Número negativo. Apenas positivo")
    
        
```
