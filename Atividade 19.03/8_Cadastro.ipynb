```python
name = input("Digite seu nome --> ")
while True:
    if not name:
        print("XX-----Nome obrigatório-----XX\n")
        name = input("Digite seu nome --> ")
    if name:
        age = int(input("Informe sua idade --> "))
        if age >= 18:
            print("Cadastro Completo")
            break
        else:
            print("Cadastro Incompleto: menor de idade")
            break
    

    
    
```
