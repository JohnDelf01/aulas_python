```python
purchase_value = int(input("Informe o valor da compra -->"))
print("Possui cupom de desconto?\n  1 para 'Sim'\n  0 para 'Não'")
discount_coupon = input()
```

```python
if purchase_value >= 100:
    if discount_coupon == "1":
        print("Cupom válido. Aplicando desconto...")
    if discount_coupon == "0":
        print("Não possui Cupom. Sem desconto.")
else:
    print("Valor insufiente para descontos.")
```
