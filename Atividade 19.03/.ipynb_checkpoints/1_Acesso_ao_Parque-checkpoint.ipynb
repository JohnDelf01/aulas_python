```python
person_age = int(input("Informe sua idade"))
have_ticket = input("Possui ticket? S/N")
```

```python
if person_age >= 12:
    if have_ticket == "s":
        print("Acesso Liberado")
    elif have_ticket == "S":
        print("Acesso Liberado")
    else:
        print("Compre um ticket")
else:
    print("Menor de idade, acesso negado")
```
