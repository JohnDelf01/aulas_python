```python
#verificação de maioridade
user_age = int(input("Informe a sua idade:"))
user_age
```

```python
nota_aluno = float(input("Informe sua nota: "))
nota_aluno
```

```python
if nota_aluno >=7:
    print("Parabéns, você foi aprovado")
elif nota_aluno >=5:
    print("Você está em recuperação!")
else:
    print("Você foi reprovado")
```

```python
#Resto da divisão por um número '%'
num = float(input("Diga um número natural: "))
num
```

```python
if num%2 == 0:
    print(f"O {num} é par")
else:
    print(f"O {num} não é par")
```

```python
num1 = float(input("Digite um primeiro valor: "))
num2 = float(input("Digite um segundo valor: "))
```

```python
if num1 > num2:
    print(f"{num1} > {num2}")
elif num1 == num2:
    print(f"{num1} = {num2}")
else:
    print(f"{num2} > {num1}")
```

```python
#  o caso, para não importar biblioteca, raiz quadrada é a mesma coisa que um número elevado a 1/"coeficiente da raiz" 
quanti = int(input("Quantas questões você quer realizar?"))


for i in range(quanti):
    print(f"\n=======Resolvendo equação {i+1}=======\n")
    
    a = float(input("Digite o valor do coeficiente de a: "))
 
    if a == 0:
        print(f"{a} é menor que zero, insira um número diferente de zero")
        break
    
    b = float(input("Digite o valor do coeficiente de b: "))
    c = float(input("Digite o valor do coeficiente de c: "))
    
    delta = b**2 - (4*a*c)

# no caso raiz quadrada = 1/2 = 0.5
    if delta < 0:
        print(f"Delta menor do que zero: {delta}")
        print("Não há raiz real")
   
    elif delta == 0:
        print(f" x'= x''={x2}")
    else:
        x1 = (-b + delta**0.5)/2*a
        x2 = (-b - delta**0.5)/2*a
        print("Resultado: Duas raízes reais. ")
        print(f"\n Delta: {delta}")
        print(f"\nx'={x1} | x''={x2}")
   
```
