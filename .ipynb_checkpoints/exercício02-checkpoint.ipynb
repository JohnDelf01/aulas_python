Exercício 6: comparar dois números

```python
num3 = float(input("Digite o primeiro número; "))
num4 = float(input("Digite o segundo número: "))

if num3 > num4:
    print(f"O primeiro número, {num3}, é o maior")
else:
     print(f"O segundo número, {num4}, é o maior")
```

Exercício 7: Classificação de idade

```python
age = int(input("Informe sua idade: "))

if age <= 12:
    print("Idade Legal: Criança")
elif age <= 17:
    print("Idade Legal: Adolescente")
else:
    print("Idade Legal: Adulto")
```

Exercício 8: Nota do aluno

```python
nota1 = float(input("Informe sua nota: "))

if nota1 >= 7:
    print("Aprovado")
elif nota1 >= 5:
    print("Recuperação")
else:
    print("Reprovado")
    
```

Exercício 9: Comparar três números

```python
n = float(input("Primeiro número: "))
n2 = float(input("Segundo número: "))
n3 = float(input("Terceiro número: "))

if n > n2 and n > n3:
    print(f"O primeiro número é o maior de todos, sendo ele: {n}")
elif n2 > n and n2 > n3:
    print(f"O segundo número é o maior de todos, sendo ele: {n2}")
else:
    print(f"O terceira número é o maior de todos, sendo ele: {n3}")
```

Exercício 10: Sistema simples de login

```python
true_user = "admin"
true_password = 1234

user = input("Digite o nome de úsuario: ")
password = int(input("Digite a senha (somente digitos): "))

if user == true_user and password == true_password:
    print("acesso permitido")
else:
    print("Acesso negado")
    
```

Desafio Extra

```python
number = float(input("Digite um número: "))

if number < 0:
    print("Número negativo")
elif number == 0:
    print("Você digitou o número 0")
else:
    print("Número positivo")
```
