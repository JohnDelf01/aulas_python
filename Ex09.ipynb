```python
n = float(input("Primeiro número: "))
n2 = float(input("Segundo número: "))
n3 = float(input("Terceiro número: "))

if n > n2 and n > n3:
    print(f"O primeiro número é o maior de todos, sendo ele: {n}")
elif n2 > n and n2 > n3:
    print(f"O segundo número é o maior de todos, sendo ele: {n2}")
else:
    print(f"O terceira número é o maior de todos, sendo ele: {n3}")
```
